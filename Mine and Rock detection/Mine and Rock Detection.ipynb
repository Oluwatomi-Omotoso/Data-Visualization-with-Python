Importing in the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from math import floor as fl
from math import ceil as cl

```

Importing in data_sets


```python
headers = [ ]
for i in range(52):
    headers.append(i)
sonar_data = pd.read_csv('./Copy of sonar data.csv', header = None)

#Viewing the data. we only get 5 rows by default.
# sonar_data.head() 


```

Data Stats

```python
# Number of columns and rows
# sonar_data.shape

# Other summaries we can get from the data(mean, percentage and standard deviation...)
# sonar_data.describe()
```

```python
# The data that helps differentiate a rock from a mine is the 60th column. 
sonar_data[60].value_counts()
# We have almost equal amounts so we can in fact make accurate predictions from this data set. Even though this isn't enough for a real world scenareo.
```

```python
# On average each column had so and so amount of rocks and so and so amount of mines.
sonar_data.groupby(60).mean()
```

Stripping apart data sets for Model Testing

```python
data = sonar_data.drop(columns = 60, axis=1)
labels = sonar_data[60]
```

Training and Test Data

```python
data_train, data_test, labels_train, labels_test = train_test_split(data,labels,test_size = 0.1, stratify = labels, random_state= 1)
```

```python
# Debugging
print(data.shape, data_train.shape, data_test.shape )


```

Model Training [Logistic regresssion]


```python
model = LogisticRegression()
```

```python
#Training the model with the data.
model.fit(data_train, labels_train)
```

Model Evaluation
 

```python
# Accuracy on training data
data_train_prediction = model.predict(data_train)
training_data_accuracy = accuracy_score(data_train_prediction,labels_train)
```

```python
print("Accuracy: ",(cl(training_data_accuracy * 100)),"%")
```

```python
# Accuracy on test data
data_test_prediction = model.predict(data_test)
testing_data_accuracy = accuracy_score(data_test_prediction,labels_test)
```

```python
print("Accuracy on test data: ",(cl(testing_data_accuracy*100)),"%")
```
